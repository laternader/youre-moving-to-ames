# Let's start some engineering

One obvious feature I wanted to incorporate was the log of the `saleprice` column in the `cleaned_train` dataset.

```python
cleaned_train['log_saleprice'] = cleaned_train['saleprice'].map(np.log)
```

Since I notice a decent feature between `totrms_abvgrd` with `bedroom_abvgrd` and `full_bath`, I will create an iteraction between them that will be included in the some select feature sets.

```python
# check for any 0s in any of the columns above
cleaned_train[(cleaned_train['totrms_abvgrd'] == 0) | (cleaned_train['bedroom_abvgr']==0) | (cleaned_train['full_bath']==0)]

# Since there are only 4 rows where it applies, then I won't drop any rows.
```

```python
cleaned_train['totrms * bedroom'] = cleaned_train['totrms_abvgrd'] * cleaned_train['bedroom_abvgr']
```

```python
# I will add up full_bath and half_bath to create total_bath, not including basement
cleaned_train['total_bath_nobsmt'] = cleaned_train['full_bath'] + cleaned_train['half_bath']
# Then I will multiply it with totrms_abvgrd
cleaned_train['totrms * baths'] = cleaned_train['totrms_abvgrd'] * cleaned_train['total_bath_nobsmt']
```

```python
# I will also create an interaction of bedrooms, bathrooms, and totrms
cleaned_train['all_rms']=cleaned_train['total_bath_nobsmt'] * cleaned_train['totrms_abvgrd'] * cleaned_train['bedroom_abvgr']
```

The same will be done to the test set for consistency sake.

```python
cleaned_test['totrms * bedroom'] = cleaned_test['totrms_abvgrd'] * cleaned_test['bedroom_abvgr']
# I will add up full_bath and half_bath to create total_bath, not including basement
cleaned_test['total_bath_nobsmt'] = cleaned_test['full_bath'] + cleaned_test['half_bath']
# Then I will multiply it with totrms_abvgrd
cleaned_test['totrms * baths'] = cleaned_test['totrms_abvgrd'] * cleaned_test['total_bath_nobsmt']
# I will also create an interaction of bedrooms, bathrooms, and totrms
cleaned_test['all_rms']=cleaned_test['total_bath_nobsmt'] * cleaned_test['totrms_abvgrd'] * cleaned_test['bedroom_abvgr']
```
