```python
# Mapping 'mc_subclass'
# Mapped ms_subclass
train['ms_subclass']=train['ms_subclass'].map({40:40, # 1 story homes
                         20:40, 
                         30:40, 
                         45:50, # 1-1.5 homes
                         50:50, 
                         60:75, # 2 story
                         70:75, 
                         75:75, 
                         180:190, # family homes
                         190:190,
                         120:160, # PUDs
                         150:160,
                         160:160,
                         90:90, # duplex
                         80:85, # split
                         85:85, 
                        })
```

```python
# neighborhood mapping
train['neighborhood']=train['neighborhood'].map({'NAmes':'NAmes',
                         'CollgCr':'CollgCr',
                         'OldTown':'OldTown',
                         'Edwards':'Edwards',
                         'Somerst':'Somerst',
                         'NridgHt':'NridgHt',
                         'Gilbert':'Gilbert',
                         'Sawyer':'Sawyer',
                         'NWAmes':'NWAmes',
                         'SawyerW':'SawyerW',
                         'Mitchel':'Mitchel',
                         'BrkSide':'BrkSide',
                         'Crawfor':'Crawfor',
                         'IDOTRR':'IDOTRR',
                         'NoRidge':'NoRidge',
                         'Timber':'Timber',
                         'StoneBr':'StoneBr',
                         'SWISU':'SWISU',
                         'ClearCr':'ClearCr',
                         'MeadowV':'MeadowV',
                         'Blmngtn':'Blmngtn',
                         'BrDale':'BrDale',
                         'NPkVill':'NPkVill',
                         'Veenker':'Veenker',
                         'Blueste':'Blueste',
                         'Greens':'Greens',
                         'GrnHill':'Timber', # not in test set so map to Timber
                         'Landmrk':'CollgCr'}) # next to college via maps
```

```python
# Mapping heating qc
train['heating_qc']=train['heating_qc'].map({'Ex':'Ex',
                                             'TA':'TA', 
                                             'Gd':'Gd', 
                                             'Fa':'Fa', 
                                             'Po':'Fa'})
```

```python
# mapping functional
train['functional']=train['functional'].map({'Typ':'Typ',
                         'Mod':'Mod',
                         'Min2':'Min',
                         'Maj1':'Maj',
                         'Min1':'Min',
                         'Sev':'Maj',
                         'Sal':'Maj',
                         'Maj2':'Maj'})
```

```python
# mapping electrical
train.electrical=train.electrical.map({'Mix':'Mix',
                    'FuseP':'Fuse',
                     'SBrkr':'Sbrkr',
                     'FuseA':'Fuse',
                     'FuseF':"Fuse"})
```

```python
# mapping kitchen_qual
train['kitchen_qual']=train['kitchen_qual'].map({'Fa':'Fa',
                          'TA':'TA',
                          'Gd':'Gd',
                          'Ex':'Ex',
                          'Po':'Fa'})
```

```python
# mapping sale_type
train['sale_type']=train['sale_type'].map({'WD ':'WD',
                       'New':'New',
                       'Con':'Con',
                       'COD':'COD',
                       'VWD':'Oth', # changed to other because not the same as WD
                       'CWD':'CWD',
                       'ConLD':'ConLD',
                       'ConLI':'ConLI',
                       'Oth':'Oth',
                       'ConLw':'ConLw'})
```

```python
# mapping garage_qual
train['garage_qual']=train['garage_qual'].map({'TA':'TA',
                          'Fa':'Fa',
                          'Gd':'Gd',
                          'Ex':'Gd',
                          'Po':'Po'})
```

```python
# mapping bsmt_cond
train['bsmt_cond']=train['bsmt_cond'].map({'TA':'Avg',
                        'Gd':'Gd',
                        'Fa':'Avg',
                        'Po':'Avg',
                        'Ex':'Gd'})
```

```python
# mapping overall_qual
train['overall_qual']=train['overall_qual'].map({6:6,
                           7:7,
                           5:5,
                           8:8,
                           10:10,
                           4:4,
                           9:9,
                           3:3,
                           2:2,
                           1:2})
```

```python
# map mas_vnr_area to fill nan with 0
train['mas_vnr_area'].fillna(0,inplace=True)
```

```python
# map the 1 row to the mean
train['total_bsmt_sf'].fillna(train['total_bsmt_sf'].mean(),inplace=True)
```

```python
# mapping nulls for garage_yr_blt and garage_area
train['garage_yr_blt'].fillna(np.round(train['garage_yr_blt'].mean()),inplace=True)

train['garage_area']=train['garage_area'].replace(0,np.nan)

train['garage_area'].fillna(np.round(train['garage_area'].mean()),inplace=True)
```

```python
# mapping 'garage_cars'
train['garage_cars'].fillna(0, inplace=True) # make it 0 because it might not even have a garage
```
